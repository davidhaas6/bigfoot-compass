```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# The columns that in the flutter data model
STORED_COLS = ['latitude', 'longitude', 'loc_estimated', 'location_details',
               'title', 'observed', 'classification', 'date', 'season','state','county',
                'moon_phase', 'humidity']
```

```python
# Load in dataframe, set sighting number as the index / id
df = pd.read_csv('sightings_with_estimates.csv')
df.set_index(df.number.astype(int), inplace=True)
df = df[STORED_COLS]

# add a creature column
df['creature'] = 'Bigfoot'
```

```python
# create the db connection
engine = create_engine('sqlite:///sightings.db', echo=False)

# drop prev sightings table
engine.execute("DROP TABLE sightings")

# write dataframe to a table
df.to_sql('sightings', con=engine)

# print table columns
engine.execute("SELECT * FROM sightings").keys()
```
